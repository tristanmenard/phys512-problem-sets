## PHYS 512 - Problem Set 5
### 1) Method of relaxation

The method of relaxation is an iterative algorithm that ensures the solution is continuous by obtaining a cell's next iteration value from the average value of the 4 adjacent current cell values. A fast way to get this average for large arrays is to take the 2-dimensional convolution of the current cell values with the following kernel:

$$ \text{kernel}= \left(\begin{matrix}0&\frac{1}{4}&0\\\frac{1}{4}&0&\frac{1}{4}\\0&\frac{1}{4}&0\end{matrix}\right) $$

The boundaries of the array of current cell values should be set to zero and prohibited from being changed.

This convolution is only correct in free space (where there is no charge). In the grid cells where there is some charge (inside the cylinder), there must additionally be a contribution to the potential of $\pi$ times the total charge within the given cell.

```python
def relaxation_method_2d(V0, rho, mask, mask_value, niter=100):
    '''
    Return the 2-dimensional method of relaxation solution for a given number of iterations.
    '''
    V = V0.copy()
    # Build the kernel for the adjacent cell averaging convolution
    kernel = np.zeros([3,3])
    kernel[1,0] = 0.25
    kernel[0,1] = 0.25
    kernel[2,1] = 0.25
    kernel[1,2] = 0.25
    # Convolve the current potential array with the kernel to get the next iteration's potential array, and repeat many times.
    # Add the charge density (in this case it is the total charge of the cell) when applicable.
    for k in range(niter):
        V = convolve2d(V, kernel, mode='same', boundary='fill', fillvalue=0) + np.pi*rho
        V[mask] = mask_value # Ensure that the boundaries are always reset to the appropriate value.
    return V
```

```python
# Number of grid cells (pixels) along an axis:
N = 100
# So, the distance between gridpoints (at the center of each grid cell) is:
step = 1/N
# Get the x and y Cartesian coords. corresponding to the gridpoints.
X, Y = np.meshgrid(np.linspace(step-1, 1-step, N), np.linspace(step-1, 1-step, N))

# Radius of the cylinder
r_cylinder = np.exp(-2)

# Initialize the N x N electric potential array.
V0 = np.zeros([N, N])
# Find the indices of the cells whose central gridpoint is within the radius of the cylinder.
idx_inside_cylinder = np.sqrt(X**2 + Y**2) <= r_cylinder
# Set the potential for gridpoints inside the cylinder to V_cylinder.
V_cylinder = np.log(r_cylinder)
V0[idx_inside_cylinder] = V_cylinder

# Mask out the boundary condition potentials (V=0 at exterior edges of the box).
mask = np.zeros([N, N], dtype='bool')
# Also mask out the manually set potentials inside the cylinder
mask[idx_inside_cylinder] = 1

# Initialize the charge density array.
rho = np.zeros([N,N])
# The charge density inside the conductive cylinder is constant (uniformally distributed) because there is no electric current.
# Let the total charge of the cylinder cross-section to be equal to 1.
# The charge of a grid cell within the cylinder is 1/(number of cells in the cylinder).
# Find the indices of the cells whose central gridpoint is within the radius of the cylinder.
idx_inside_cylinder = np.sqrt(X**2 + Y**2) <= r_cylinder
n_inside_cylinder = np.sum(idx_inside_cylinder) # is the number of grid cells in the cylinder
rho[idx_inside_cylinder] = 1/n_inside_cylinder # is the total charge of each grid cell
```

```python
plt.title('Charge density')
plt.imshow(rho, extent=[-1,1,-1,1], origin='lower')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([-1,1,-1,1])
plt.colorbar()
plt.show()
```

```python
# Get the 2d method of relaxation potentials for different numbers of iterations.
V_100 = relaxation_method_2d(V0, rho, mask, V_cylinder, 100)
V_1000 = relaxation_method_2d(V0, rho, mask, V_cylinder, 1000)
V_10000 = relaxation_method_2d(V0, rho, mask, V_cylinder, 10000)
```

### 2) Conjugate gradient method

```python
def RHS_2d(pot, mask):
    '''
    Returns the matrix of the boundary conditions effect on each grid cell in a 2-dimensional linear problem given by Ax=b.
    '''
    mat = np.zeros(pot.shape)
    # Roll the matrix to each side, up and down by 1 cell to add up all boundary conditions from adjacent cells.
    mat[:,:-1] = mat[:,:-1] + pot[:,1:]
    mat[:,1:] = mat[:,1:] + pot[:,:-1]
    mat[:-1,:] = mat[:-1,:] + pot[1:,:]
    mat[1:,:] = mat[1:,:] + pot[:-1,:]
    # Reset the masked cells to zero.
    mat[mask] = 0
    return mat

def Ax_2d(mat, mask, copy=False):
    '''
    Returns the coefficient matrix A multiplied by the vector of unknowns x for a 2-dimensional linear problem given by Ax=b.
    '''
    if copy:
        mat = mat.copy()
    mat[mask] = 0
    mm = 4*mat
    # Roll the matrix side-to-side, up and down to add up all contibutions from adjacent cells.
    mm[:,:-1] = mm[:,:-1] - mat[:,1:]
    mm[:,1:] = mm[:,1:] - mat[:,:-1]
    mm[1:,:] = mm[1:,:] - mat[:-1,:]
    mm[:-1,:] = mm[:-1,:] - mat[1:,:]
    mm[mask] = 0
    return mm

def CG(rhs, x0, mask, niter=100, tolerance=0.00001, verbose=True):
    '''
    Returns the solution x to the linear problem given by Ax=b using the conjugate gradient algorithm.
        x0: the initial guess for the solution
    '''
    Ax = Ax_2d(x0, mask)
    r = rhs - Ax # the residual
    p = r.copy()
    x = x0.copy()
    rsqr = np.sum(r*r)
#     print('starting rsqr is ',rsqr)
    for k in range(niter):
        Ap = Ax_2d(p, mask)
        alpha = np.sum(r*r)/np.sum(Ap*p)
        x = x + alpha*p
        r = r - alpha*Ap
        rsqr_new = np.sum(r*r)
        
        # Break out of the loop and return x if value of the new residual is less than some tolerance in every grid cell
        if np.sqrt(rsqr_new) < tolerance*len(r):
            if verbose:
                print('Conjugate gradient method converged after {:d} iterations.'.format(k))
            break
        elif k == niter-1:
            print('Conjugate gradient method did not converge after {:d} iterations'.format(k+1))
            break
        
        beta = rsqr_new/rsqr
        p = r + beta*p
        rsqr = rsqr_new
#     print('final rsqr is ', rsqr)
    return x
```

```python
# Number of grid cells (pixels) along an axis:
N = 100
# So, the distance between gridpoints (at the center of each grid cell) is:
step = 1/N
# Get the x and y Cartesian coords. corresponding to the gridpoints.
X, Y = np.meshgrid(np.linspace(step-1, 1-step, N), np.linspace(step-1, 1-step, N))

# Radius of the cylinder
r_cylinder = np.exp(-2)

# Initialize the N x N electric potential array.
V = np.zeros([N, N])
# Set the potential for gridpoints inside the cylinder to V_cylinder.
V_cylinder = np.log(r_cylinder)
V[np.sqrt(X**2 + Y**2) <= r_cylinder] = V_cylinder

# Mask out the boundary condition potentials (V=0 at exterior edges of the box).
mask = np.zeros([N, N], dtype='bool')
mask[0,:] = 1
mask[:,0] = 1
mask[-1,:] = 1
mask[:,-1] = 1
# Also mask out the manually set potentials inside the cylinder
mask[np.sqrt(X**2 + Y**2) <= r_cylinder] = 1
```

```python
# Do the conjugate gradient algorithm for the problem that was set up!
rhs = RHS_2d(V, mask)
x = CG(rhs, 0*rhs, mask, niter=150) # x gives the converged electric potential solution V for each grid cell

# x has the masked out regions set to zero currently, so add those potential back in
x[mask] = V[mask]

# Plot the final solution to the problem
plt.title('Conjugate gradient method: {:d}x{:d} resolution'.format(N,N), y=1.02)
im = plt.imshow(x.T, origin='lower', extent=[-1,1,-1,1], cmap='viridis_r')
cs = plt.contour(X, Y, x.T, [-2,-1.5,-1,-0.5,0], colors='k', linewidths=1)
plt.clabel(cs, fmt='%2.1f')
cbar = plt.colorbar(im)
cbar.add_lines(cs)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

For the same resolution as in 1), the conjugate gradient algorithm converged in 121 iterations as opposed to the method of relaxation that took $\sim 10^4$ iterations to converge. Put another way, the conjugate gradient method is over 100 times faster that the relaxation method, and this allows us to spend more time resolving even finer details...

### 3) Conjugate gradient method on a range of resolutions

Let's try for a higher resolution solution with the conjugate gradient method.

```python
# Using the conjugate gradient solve from 2): (copy-pasted from above -- changed the number of pixels N from 100 to 1024, added timing mechanism)
# Number of grid cells (pixels) along an axis:
N = 1024
# So, the distance between gridpoints (at the center of each grid cell) is:
step = 1/N
# Get the x and y Cartesian coords. corresponding to the gridpoints.
X, Y = np.meshgrid(np.linspace(step-1, 1-step, N), np.linspace(step-1, 1-step, N))

# Radius of the cylinder
r_cylinder = np.exp(-2)

# Initialize the N x N electric potential array.
V = np.zeros([N, N])
# Set the potential for gridpoints inside the cylinder to V_cylinder.
V_cylinder = np.log(r_cylinder)
V[np.sqrt(X**2 + Y**2) <= r_cylinder] = V_cylinder

# Mask out the boundary condition potentials (V=0 at exterior edges of the box).
mask = np.zeros([N, N], dtype='bool')
mask[0,:] = 1
mask[:,0] = 1
mask[-1,:] = 1
mask[:,-1] = 1
# Also mask out the manually set potentials inside the cylinder
mask[np.sqrt(X**2 + Y**2) <= r_cylinder] = 1

rhs = RHS_2d(V, mask)

# How long does it take to find the converged conjugate gradient solution at this high resolution?
t1 = time.time()
x = CG(rhs, 0*rhs, mask, niter=1000)
t2 = time.time()
print('Conjugate gradient solver from 2) converged in ', t2-t1, ' s.')

x[mask] = V[mask] # add the masked values back in for the final solution

# Plot the final solution to the problem
plt.title('Conjugate gradient method: {:d}x{:d} resolution'.format(N,N), y=1.02 )
im = plt.imshow(x.T, origin='lower', extent=[-1,1,-1,1], cmap='viridis_r')
cs = plt.contour(X, Y, x.T, [-2,-1.5,-1,-0.5,0], colors='k')
plt.clabel(cs, fmt='%2.1f')
cbar = plt.colorbar(im)
cbar.add_lines(cs)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

It takes approximately 83 seconds and 946 iterations to the converge to a high resolution (1024 pixel) solution with the conjugate gradient solver from 2). 

Now, let's try speeding this up by solving a coarser resolution problem, then interpolating the result into the guess $x_0$ for the high resolution problem.

```python
def resolve_less(mat):
    '''
    Return an interpolated array that is two times coarser in resolution than the input array.
    '''
    assert np.all(mat >= 0) or np.all(mat <= 0), 'Array mixes positive and negative values and may produce unexpected results.'
    mm=np.zeros([int(mat.shape[0]/2),int(mat.shape[1]/2)], dtype=mat.dtype)
    if np.all(mat >= 0):
        mm=np.maximum(mm,mat[::2,::2])
        mm=np.maximum(mm,mat[::2,1::2])
        mm=np.maximum(mm,mat[1::2,::2])
        mm=np.maximum(mm,mat[1::2,1::2])
    else:    
        mm=np.minimum(mm,mat[::2,::2])
        mm=np.minimum(mm,mat[::2,1::2])
        mm=np.minimum(mm,mat[1::2,::2])
        mm=np.minimum(mm,mat[1::2,1::2])
    return mm
    
def resolve_more(mat):
    '''
    Return an interpolated array that is two times finer in resolution than the input array.
    '''
    mm=np.zeros([int(mat.shape[0]*2),int(mat.shape[1]*2)],dtype=mat.dtype)
    mm[::2,::2]=mat
    mm[::2,1::2]=mat
    mm[1::2,::2]=mat
    mm[1::2,1::2]=mat
    return mm
```

```python
# Initialize arrays to hold masks, boundary conditions, rhs, and solutions for the different resolutions.
npass = 6
all_masks = [None]*npass
all_V = [None]*npass
all_rhs = [None]*npass
all_x = [None]*npass
# Already have the highest resolution mask and boundary conditions.
all_masks[0] = mask
all_V[0] = V
# Get the masks and boundary conditions for each lower resolution problem.
for i in range(1, npass):
    all_masks[i] = resolve_less(all_masks[i-1])
    all_V[i] = resolve_less(all_V[i-1])
```

```python
# Plot the masks (to verify they are correct)
f, ax = plt.subplots(1, npass, figsize=(14,6), sharey=True)

ax[0].set_ylabel('y')
for i in range(npass):
    ax[i].set_title('{:d} px'.format(int(2*N/(2**(i+1)))))
    im = ax[i].imshow(all_masks[i], origin='lower', extent=[-1,1,-1,1], cmap='viridis_r')
    ax[i].set_xlabel('x')
     
plt.suptitle('Full range of mask resolutions', y=0.74)
plt.tight_layout()
plt.show()

# Plot the initial potentials (to verify they are correct)
f, ax = plt.subplots(1, npass, figsize=(14,6), sharey=True)

ax[0].set_ylabel('y')
for i in range(npass):
    ax[i].set_title('{:d} px'.format(int(2*N/(2**(i+1)))))
    im = ax[i].imshow(all_V[i], origin='lower', extent=[-1,1,-1,1], cmap='viridis_r')
    ax[i].set_xlabel('x')
     
plt.suptitle('Full range of V boundary resolutions', y=0.74)
plt.tight_layout()
plt.show()
```

```python
# How long does it take to get the highest resolution solution this way?
t1 = time.time()
# Make the lowest resolution map (i=-1)
all_rhs[-1] = RHS_2d(all_V[-1], all_masks[-1])
all_x[-1] = CG(all_rhs[-1], 0*all_rhs[-1], all_masks[-1])
all_x[-1][all_masks[-1]] = V_cylinder # add the masked values back in to the low res solution

# Increase the resolution of x, set it equal to the initial guess for the next higher resolution, get the new x solution, and repeat until the desired resolution is acheived.
for i in range(npass-2, -1, -1):
        all_rhs[i] = RHS_2d(all_V[i], all_masks[i])
        x0 = resolve_more(all_x[i+1])
        all_x[i] = CG(all_rhs[i], x0, all_masks[i], niter=150)
        all_x[i][all_masks[i]] = V_cylinder
t2 = time.time()
print('Resolution-varying conjugate gradient solver converged in', t2-t1, ' s.')
```

```python
f, ax = plt.subplots(1, 6, figsize=(16,6), sharey=True)

ax[0].set_ylabel('y')
for i in range(npass):
    ax[-i-1].set_title('{:d} px'.format(int(2*N/(2**(i+1)))))
    im = ax[-i-1].imshow(all_x[i], origin='lower', extent=[-1,1,-1,1], vmax=0, cmap='viridis_r')
    ax[-i-1].set_xlabel('x')
     
plt.suptitle('Conjugate gradient solutions for all resolutions', y=0.74)
f.colorbar(im, ax=ax, fraction=0.0071, pad=0.01)
# plt.tight_layout()
plt.show()
```

It only takes around 10 seconds now to get to a solution to the same threshold and at the same high resolution as previously. Also, the previous attempt required 946 iterations to arrive at a converged solution. Now, it only takes 510 iterations. So, this is a more than a nearly 10 times increase in speed alongside a 2 times decrease in computational intensity.

### 4) A bump in the wire

```python
# Number of grid cells (pixels) along an axis:
N = 1024
# So, the distance between gridpoints (at the center of each grid cell) is:
step = 1/N
# Get the x and y Cartesian coords. corresponding to the gridpoints.
X, Y = np.meshgrid(np.linspace(step-1, 1-step, N), np.linspace(step-1, 1-step, N))

# Radius of the cylinder
r_cylinder = np.exp(-2)
# Radius of the bump
r_bump = 0.1*r_cylinder

# Initialize the N x N electric potential array.
V = np.zeros([N, N])
# Set the potential for gridpoints inside the cylinder to V_cylinder.
V_cylinder = np.log(r_cylinder)
V[np.sqrt(X**2 + Y**2) <= r_cylinder] = V_cylinder
# Set the potential inside the bump at (x=r_cylinder, y=0) to V_cylinder.
V[np.sqrt((X-r_cylinder)**2 + Y**2) <= r_bump] = V_cylinder

# Mask out the boundary condition potentials (V=0 at exterior edges of the box).
mask = np.zeros([N, N], dtype='bool')
mask[0,:] = 1
mask[:,0] = 1
mask[-1,:] = 1
mask[:,-1] = 1
# Also mask out the manually set potentials inside the cylinder and the bump
mask[np.sqrt(X**2 + Y**2) <= r_cylinder] = 1 # cylinder
mask[np.sqrt((X-r_cylinder)**2 + Y**2) <= r_bump] = 1 # bump
```

```python
npass = 4
all_masks = [None]*npass
all_V = [None]*npass
all_rhs = [None]*npass
all_x = [None]*npass

all_masks[0] = mask
all_V[0] = V

for i in range(1, npass):
    all_masks[i] = resolve_less(all_masks[i-1])
    all_V[i] = resolve_less(all_V[i-1])
    
# Make the lowest resolution map (i=-1)
all_rhs[-1] = RHS_2d(all_V[-1], all_masks[-1])
all_x[-1] = CG(all_rhs[-1], 0*all_rhs[-1], all_masks[-1], niter=200)
all_x[-1][all_masks[-1]] = all_V[-1][all_masks[-1]]

# Increase the resolution of x, set it equal to the initial guess for the next higher resolution, get the new x solution, and repeat until the desired resolution is acheived.
for i in range(npass-2, -1, -1):
        all_rhs[i] = RHS_2d(all_V[i], all_masks[i])
        x0 = resolve_more(all_x[i+1])
        all_x[i] = CG(all_rhs[i], x0, all_masks[i])
        all_x[i][all_masks[i]] = all_V[i][all_masks[i]]
  
f, ax = plt.subplots(1, npass, figsize=(16,5), sharey=True)
ax[0].set_ylabel('y')
for i in range(npass):
    ax[-i-1].set_title('{:d} px'.format(int(2*N/(2**(i+1)))))
    im = ax[-i-1].imshow(all_x[i], origin='lower', extent=[-1,1,-1,1], vmin=np.log(r_cylinder), vmax=0, cmap='viridis_r')
    ax[-i-1].set_xlabel('x')   
plt.suptitle('Conjugate gradient solutions (in free space) for all resolutions', y=0.88)
cbar = f.colorbar(im, ax=ax, fraction=0.01)
plt.show()
```

```python
# Add the potential of the wire and bump back in
all_x[0][all_masks[0]] = V[all_masks[0]]
all_x[0][all_masks[0]] = V[all_masks[0]]

# Plot the electric potential at the highest resolution
plt.figure(figsize=(8,6))
plt.title('Electric potential')
im = plt.imshow(all_x[0], origin='lower', extent=[-1,1,-1,1], cmap='viridis_r')
cs = plt.contour(X, Y, all_x[0], [-2,-1.5,-1,-0.5], colors='k', linewidths=1)
plt.clabel(cs, fmt='%2.1f')
cbar = plt.colorbar(im)
cbar.add_lines(cs)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Get the x and y gradients to find the plot the electric field
u = np.gradient(all_x[0])[1]
v = np.gradient(all_x[0])[0]

# Plot the electric field magnitude at the highest potential
plt.figure(figsize=(8,6))
plt.title('Electric field')
plt.imshow(np.sqrt(u**2+v**2), origin='lower', extent=[-1,1,-1,1])
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
res = 4 # plot 1 arrow for every res^2 of grid cells

plt.figure(figsize=(10,6))
plt.quiver(X[::res, ::res], Y[::res, ::res], u[::res, ::res], v[::res, ::res], scale=0.6)
plt.imshow(np.sqrt(u**2+v**2), extent=[-1,1,-1,1], origin='lower')
plt.axis([-0.2,0.3,-0.2,0.2])
plt.colorbar()
plt.show()
```

Compared to the far side from the bump, the magnitude of the electric field is higher near the tip of the bump and it is lower where the bump meets the wire. Additionally, the electric field is more scattered around the bump.

### 5) Temperature in a box as a function of time

There are a couple different ways to set up the boundary conditions here. First, I will look at the case, where the temperature rises by 1 in every time step along the left wall and all other walls are kept at $T=0$. Essentially the three other wall are acting as heat sinks in this case.

```python
# Number of interior gridpoints:
N = 256
# The temperature of the left wall (x=0) will be raised by dT every time step.
dT = 1
# Number of time steps:
nstep = 50

# Initialize the N x N temperature array.
# Start all grid cells at T=0, then update the temperature along the left wall for every time step.
T = np.zeros([N,N])

# Initialize the mask array
mask = np.zeros([N,N], dtype='bool')
mask[0,:] = 1
mask[:,0] = 1
mask[-1,:] = 1
mask[:,-1] = 1

# Initialize an array to hold onto the solution along the center line at every time step
x_all = np.zeros([nstep+1, N])

T[:,0] += dT # update the left wall's temperature
rhs = RHS_2d(T, mask)
x = CG(rhs, 0*rhs, mask, niter=1000, verbose=False)
x[mask] = T[mask]
x_all[0] = x[int(N/2)] # hold onto the solution along the center line (x-axis)

for i in range(1,nstep+1):
    x0 = x # use the last timestep as the guess for the next one
    T[:,0] += dT
    rhs = RHS_2d(T, mask)
    x = CG(rhs, x0, mask, niter=1000, verbose=False)
    x[mask] = T[mask]
    x_all[i] = x[int(N/2)]  # hold onto the solution along the center line (x-axis)

plt.title('Temperature along center line as time increases')
plt.plot(np.linspace(0,1,N), x_all[::int(nstep/10)].T)
plt.ylabel('Temperature')
plt.xlabel('x (along the center line)')
plt.xlim(0,1)
plt.ylim(0,nstep*dT+dT)
plt.legend(['t={:d}'.format(int(i*5)) for i in range(int(nstep/5+1))])
plt.grid()
plt.show()
```

Another possible choice of boundary conditions would be to allow the three other walls to absorb the heat from adjacent cells and re-emit it during the next iteration. In this case, the temperature of the three walls will act as a sort of feedback mechanism and increase the total temperature within the box.

```python
# Number of interior gridpoints:
N = 200
# The temperature of the left wall (x=0) will be raised by dT every time step.
dT = 1
# Number of time steps:
nstep = 50

# Initialize the N x N temperature array.
# Start all grid cells at T=0, then update the temperature along the left wall for every time step.
T = np.zeros([N,N])

# Initialize the mask array
mask = np.zeros([N,N], dtype='bool')
mask[0,:] = 1
mask[:,0] = 1
mask[-1,:] = 1
mask[:,-1] = 1

# Initialize an array to hold onto the solution along the center line at every time step
x_all = np.zeros([nstep+1, N])

T[:,0] += dT # update the left wall's temperature
rhs = RHS_2d(T, mask)
x = CG(rhs, 0*rhs, mask, niter=1000, verbose=False)
x[mask] = T[mask]
x_all[0] = x[int(N/2)]

for i in range(1, nstep+1):
    x0 = x # use the last timestep as the guess for the next one
    T[:,0] += dT
    T[0,:] = x[1,:]
    T[-1,:] = x[-2,:]
    T[:,-1] = x[:,-2]
    rhs = RHS_2d(T, mask)
    x = CG(rhs, x0, mask, niter=1000, verbose=False)
    x[mask] = T[mask]
    x_all[i] = x[int(N/2)]

plt.title('Temperature along center line as time increases')
plt.plot(np.linspace(0,1,N), x_all[::int(nstep/10)].T)
plt.ylabel('Temperature')
plt.xlabel('x (along the center line)')
plt.xlim(0,1)
plt.ylim(0,nstep*dT+dT)
plt.legend(['t={:d}'.format(int(i*5)) for i in range(int(nstep/5+1))])
plt.grid()
plt.show()
```

In this case, the system may behave differently depending on the choice of the constant dT and the number of grid cells (pixels). If there the number of grid cells is large, then the heat will take longer to transfer from one side of the box to the other. However, if the constant dT is small, then the heat will be able to reach the three feedback wall before the temperature of the other wall changes by too much. Therefore, in the small dT regime with a small number of pixels, the feedback walls will have a strong effect on the state of the box, and allow the temperature difference between the one wall and its opposing feedback wall to reach an equilibrium state. In the large number of pixels and large dT state regime, no equilibrium temperature difference will be reached.
