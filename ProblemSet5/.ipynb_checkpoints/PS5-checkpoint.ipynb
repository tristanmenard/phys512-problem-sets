```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import DivergingNorm
```

## PHYS 512 - Problem Set 5
### 1) Method of relaxation

```python
# ANALYTIC SOLUTION
charge_density = 1
e0 = 8.85e-12
xmin = -5; xmax = 5
ymin = -5; ymax = 5
N = 100
xx = np.linspace(xmin, xmax, N)
yy = np.linspace(ymin, ymax, N)
x, y = np.meshgrid(xx, yy)
r = np.sqrt(x**2 + y**2)
V = np.log(r)

f, ax = plt.subplots()
ax.set_title(r'Analytic potential: $V(r) = \frac{\lambda}{2\pi\epsilon_0}\log(r)$', pad=10)
ax.set_aspect('equal', 'box')
# c = ax.contourf(x, y, V, 100, cmap='seismic', norm=DivergingNorm(0))
c = ax.imshow(V, extent=[xmin,xmax,ymin,ymax], cmap='seismic', norm=DivergingNorm(0), interpolation=None)
plt.colorbar(c)
plt.show()
```

```python
gaussian2D = -6*np.exp(-(0.5*x**2/0.5**2+0.5*y**2/0.5**2)) + 2

f, ax = plt.subplots()
ax.set_aspect('equal', 'box')
# c = ax.contourf(x, y, V, 100, cmap='seismic', norm=DivergingNorm(0))
c = ax.imshow(gaussian2D, extent=[xmin,xmax,ymin,ymax], cmap='seismic', norm=DivergingNorm(0), interpolation=None)
plt.colorbar(c)
plt.show()
```

```python
theta = np.linspace(0,2*np.pi,100)
r_cylinder = 1
x_cylinder = r_cylinder*np.cos(theta)
y_cylinder = r_cylinder*np.sin(theta)

i_cylinder = np.digitize(x_cylinder, xx)
j_cylinder = np.digitize(y_cylinder, yy)

V_cylinder = np.zeros([len(x), len(y)])
V_cylinder[i_cylinder,j_cylinder] = 1

plt.imshow(V_cylinder, extent=[xmin,xmax,ymin,ymax])
plt.show()
```

```python
# Use 2d gaussian function as guess for the potential.
# Set boundary conditions: V = 0 at edges, V = 5 at center (infinitessimally small cylinder)
V = gaussian2D.copy()
V[0,:] = 0
V[:,0] = 0
V[len(y)-1,:] = 0
V[:,len(x)-1] = 0
# V = np.zeros([len(y), len(x)])
V[i_cylinder,j_cylinder] = 1

f, ax = plt.subplots()
# ax.set_aspect('equal', 'box')
# c = ax.contourf(x, y, V, 100, cmap='seismic', norm=DivergingNorm(0))
c = ax.imshow(V-gaussian2D, extent=[xmin,xmax,ymin,ymax], cmap='seismic', norm=DivergingNorm(0), interpolation=None)
plt.grid()
plt.colorbar(c)
plt.show()

niter = 1
for n in range(niter):
    for i in range(1, len(y)-1):
        for j in range(1, len(x)-1):
            if any(i == i_cylinder) and any(j == j_cylinder):
                pass
            else:
                V[i,j] = (V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1])/4

plt.imshow(V, extent=[xmin,xmax,ymin,ymax], cmap='seismic', norm=DivergingNorm(0), interpolation=None)
plt.colorbar()
plt.show()

niter = 10
for n in range(niter):
    for i in range(1, len(y)-1):
        for j in range(1, len(x)-1):
            for k in range(len(i_cylinder)):
                if i == i_cylinder[k] and j == j_cylinder[k]:
                    pass
                else:
                    V[i,j] = (V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1])/4

plt.imshow(V, extent=[xmin,xmax,ymin,ymax], cmap='seismic', norm=DivergingNorm(0), interpolation=None)
plt.colorbar()
plt.show()
```

```python
V = np.zeros([len(y), len(x)])
V[i_cylinder,j_cylinder] = 1
V0 = V.copy()

for i in range(1, len(y)-1):
    for j in range(1, len(x)-1):
        for k in range(len(i_cylinder)):
            if i == i_cylinder[k] and j == j_cylinder[k]:
                V[i,j] = -1

plt.imshow(V0-V)
plt.colorbar()
plt.show()
```
