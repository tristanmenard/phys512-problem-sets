### 1) Method of relaxation

Poisson's Equation in electrostatics:
$$ \nabla^2 V = \dfrac{-\rho}{\epsilon_0} \text{, where $V$ is the potential and $\rho$ is the charge density}. $$

In this case, the charge is held in a cylindrical conductor

```python
def relaxation_method(V0, h, niter):
    V = V0.copy()
    # Ensure that the initial boundary conditions are held over all iterations. (only loop from V[1] to V[len(r)-1]).
    for n in range(niter):
        for i in range(1, len(r)-1):
            V[i] = 0.5*(V[i-1] + V[i+1] - h**2*V[i])
    return V
```

```python
# ANALYTIC SOLUTION 1d
charge_density = 1
e0 = 8.85e-12
rmin = np.exp(-4)
rmax = 1
N = 100
r = np.linspace(rmin, rmax, N)
V_analytic = np.log(r) # dividing out the constants

# METHOD OF RELAXATION 1d
# Set initial potential everywhere. Cylinder is at potential V=-4 and box edges are at potential V=0
V0 = np.zeros(len(r))
V0[0] = np.log(rmin)

# Iterate using the method of relaxation:
# Let h be the grid spacing
h = r[1]-r[0]

V100 = relaxation_method(V0, h, int(1e2))
V1000 = relaxation_method(V0, h, int(1e3))
V10000 = relaxation_method(V0, h, int(1e4))

plt.figure(figsize=(8,5))
plt.title('Potentials')
plt.plot(r, V_analytic, 'k--', label=r'analytic: $V(r) = \dfrac{\lambda}{2\pi\epsilon_0}\log(r)$')
plt.plot(r, V100, label=r'method of relaxation [$10^2$ iterations]')
plt.plot(r, V1000, label=r'method of relaxation [$10^3$ iterations]')
plt.plot(r, V10000, label=r'method of relaxation [$10^4$ iterations]')
plt.ylabel(r'$V(r)/\left(\dfrac{\lambda}{2\pi\epsilon_0}\right)$')
plt.xlabel('r')
plt.legend()
plt.grid()
plt.show()
```

The method of relaxation takes $\sim 10^4$ iterations to converge. The converged potential as a function of r does not match the analytic potential function. The discrepancy occurs because the method of relaxation essentially smooths out the potential until it is a nearly straight line.

### 2) Conjugate gradient

### 5) Temperature

```python
def A_matrix(N):
    # Returns the A matrix for the conjugate gradient method (Ax=b)
    A = 4*np.eye(N**2,N**2) # Initialize the N^2 by N^2 A matrix
    row = 0
    for i in range(N):
        for j in range(N):
            if j+1 < N:
                A[row, N*i+j+1] = -1
            if j > 0:
                A[row, N*i+j-1] = -1
            if i+1 < N:
                A[row, N*(i+1)+j] = -1
            if i > 0:
                A[row, N*(i-1)+j] = -1
            row += 1
    return A

def b_vector(N, leftwall, rightwall, upperwall, lowerwall):
    # Returns the b vector for the conjugate gradient method (Ax=b).
    # b is a vector whose elements are the boundary values affecting each element in x
    b = np.zeros(N**2)
    for i in range(N):
        for j in range(N):
            if i == 0 and j == 0: # top left corner
                b[0] = leftwall[j] + upperwall[j]
            elif i == 0 and j == N-1: # top right corner
                b[N*i+j] = rightwall[j] + upperwall[j]
            elif i == N-1 and j == 0: # bottom left corner
                b[N*i+j] = leftwall[j] + lowerwall[j]
            elif i == N-1 and j == N-1: # bottom right corner
                b[N*i+j] = lowerwall[j] + rightwall[j]   
            elif i == 0: # top side
                b[N*i+j] = upperwall[j]
            elif j == 0: # left side
                b[N*i+j] = leftwall[j] 
            elif i == N-1: # bottom side
                b[N*i+j] = lowerwall[j]
            elif j == N-1: # right side
                b[N*i+j] = rightwall[j]
    return b

def conjgrad(A, b, N, niter=100, tolerance=0.001):
    # Returns the value of 
    x = np.zeros(N**2)
    r = b
    d = r
    for i in range(niter):
        v = np.dot(A,d)
        alpha = np.dot(r,r)/np.dot(d, v)
        x = x + alpha*d
        rnew = r - alpha*v
        if sum(abs(r)) < tolerance:
            print('Conjugate gradient method converged after {:d} iterations.'.format(i))
            break
        elif i == niter-1:
            print('Conjugate gradient method did not converge after {:d} iterations'.format(i))
            break
        beta = np.dot(rnew, rnew)/np.dot(r,r)
        d = rnew + beta*d
        r = rnew

    return x
```

```python
# Number of interior gridpoints:
N = 60
# The temperature of the left wall (x=0) will be raised by dT every time step.
dT = 1
# Number of time steps:
nstep = 10

# Boundary conditions
leftwall = np.zeros(N) # will update the left wall temperature before every time step
rightwall = np.zeros(N) # the other walls will stay at T=0 for all time
upperwall = np.zeros(N)
lowerwall = np.zeros(N)

# Get the A matrix for the conjugate gradient method.
A = A_matrix(N)

# Initialize an array to hold onto the solution along the center line at every time step
x_all = np.zeros([nstep, N])

for i in range(nstep):
    leftwall += dT # update the left wall's temperature
    b = b_vector(N, leftwall, rightwall, upperwall, lowerwall)
    x = conjgrad(A, b, N, niter=200)
    x_along_center = x[int(N*N/2):int(N*N/2+N)]
    x_all[i,:] = x_along_center

plt.title('Temperature along center line as time increases')
plt.plot(np.linspace(0,1,N), x_all.T)
plt.ylabel('Temperature')
plt.xlabel('x (along the center line)')
plt.xlim(0,1)
plt.legend(['t={:d}'.format(i) for i in range(1,N+1)])
plt.grid()
plt.show()

# plt.imshow(np.reshape(x, (N,N)))
# plt.show()

# # New boundary conditions at next time step
# leftwall = np.ones(N) + dT
# rightwall = np.zeros(N)
# upperwall = np.zeros(N)
# lowerwall = np.zeros(N)
```

I chose to keep the 3 other walls at $T=0$ for all time steps. However, other boundary choices are possible. For example, I could instead choose to let those boundaries be "reflective", meaining I allow them to take on the temperature of the nearest gridpoint. This would allow a feedback mechanism that adds more heat into the box, and therefore would result in an entirely different regime of solutions with some type of modes.
