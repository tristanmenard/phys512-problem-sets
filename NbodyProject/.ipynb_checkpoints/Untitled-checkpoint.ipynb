```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import fftconvolve
import time
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
def  get_particles_per_bin(x_particles, y_particles, nbins, extent):
    # Define the bin edges.
    xmin, xmax, ymin, ymax = extent
    xedges = np.linspace(xmin, xmax, nbins+1)
    yedges = np.linspace(ymin, ymax, nbins+1)
    
    # Get the indices of the bin array corresponding to the location of the particles.
    x_idx = np.digitize(x_particles, xedges)-1 # !!! PROBLEM HERE: index can be -1 if x_particle is < the 1st bin edge
    y_idx = np.digitize(y_particles, yedges)-1 # !!! ANOTHER PROBLEM: all particles at x_particle > last bin edge are lumped together into the last bin 
    
    # Initialize an array for the number of particles in each bin.
    particles_per_bin = np.zeros([nbins, nbins])
    # Add up the number of particles in each bin.
    for i in range(len(x_idx)):
        if x_idx[i] == nbins:
            x_idx[i] = nbins-1
        if y_idx[i] == nbins:
            y_idx[i] = nbins-1
        particles_per_bin[y_idx[i],x_idx[i]] += 1
    
    return particles_per_bin, xedges, yedges, np.column_stack((x_idx, y_idx))

def get_potential(density, x, y, soft=0.01):
    # Convolve the density array with the Green's function: G(r) = 1/abs(r)
    X, Y = np.meshgrid(x, y)
    rsqr = X**2 + Y**2
    rsqr[rsqr < soft**2] = soft**2
    green = 1/np.sqrt(rsqr)
    
    pot = fftconvolve(density, green, mode='same')
    return pot
    # Get the 2-dimensional Fourier transform of the density array.
#     density_ft = np.fft.fft2(density*win2d)
    
    # Build the Fourier transform of the finite Laplacian.
#     kx = 1/x
#     ky = 1/y
#     KX, KY = np.meshgrid(kx, ky)
#     green_ft = 1/(np.sin(np.pi*KX/len(x))**2 + np.sin(np.pi*KY/len(y))**2)
    # Convolve in Fourier space, transform back to regular space, and return the result.
#     return np.fft.ifft2(density_ft*green_ft)

def destroy_particles(x, y, vx, vy, xmin, xmax, ymin, ymax):
    # Destroy particles that have escaped the box's bounds!
    # Find and destroy particles that are outside the allowed x range.
    idx_outside_xrange = (xmin <= x)&(x < xmax)
    x_tmp = x[idx_outside_xrange]
    y_tmp = y[idx_outside_xrange]
    vx_tmp = vx[idx_outside_xrange]
    vy_tmp = vy[idx_outside_xrange]
    # Find and destroy particles that are outside the allowed y range.
    idx_outside_yrange = (ymin <= y_tmp)&(y_tmp < ymax)
    x_new = x_tmp[idx_outside_yrange]
    y_new = y_tmp[idx_outside_yrange]
    vx_new = vx_tmp[idx_outside_yrange]
    vy_new = vy_tmp[idx_outside_yrange]
    print('Destroyed {:d} particles that were out of bounds!'.format(len(x)-len(x_new))) 
    return x_new, y_new, vx_new, y_new
```

```python
# Timestep
dt = 0.1

# Number of particles:
n = 1000
# Distribute the partciles randomly in space.
x_particle = np.random.randn(n)
y_particle = np.random.randn(n)
# x_particle = np.array([1.05,1.])
# y_particle = np.array([1.,1.])
# Start all particles at rest.
vx = np.zeros(n)
vy = np.zeros(n)

# Resolution of grid (number of pixels):
npix = 127
xmin = -5; xmax = 5; ymin = -5; ymax = 5
extent = [xmin,xmax,ymin,ymax]

# Destroy particles that are outside the box.
# x_particle, y_particle, vx, vy = destroy_particles(x_particle, y_particle, vx, vy, xmin, xmax, ymin, ymax)

# Get the density function:
density, xedges, yedges, bin_idx = get_particles_per_bin(x_particle, y_particle, npix, extent)

# Gridpoints are at center of each grid cell.
x_gridpoints = (xedges[1:]+xedges[:-1])/2
y_gridpoints = (yedges[1:]+yedges[:-1])/2
X, Y = np.meshgrid(x_gridpoints, y_gridpoints)

# Plot the density:
plt.imshow(density, origin='lower', extent=extent)
plt.colorbar()
plt.show()

# Get the potential:
pot = get_potential(density, x_gridpoints, y_gridpoints)

# Plot the potential:
plt.imshow(np.abs(pot), origin='lower', extent=extent)
plt.colorbar()
plt.show()

# Get the force acting on each grid cell from the gradient of the potential:
grad_pot = np.gradient(np.abs(pot))
Fx = grad_pot[1]
Fy = grad_pot[0]

plt.figure(figsize=(12,8))
plt.imshow(np.sqrt(Fx**2+Fy**2), origin='lower', extent=extent)
plt.colorbar()
plt.quiver(x_gridpoints, y_gridpoints, Fx, Fy)
plt.show()
```

```python
nstep = 200
x_all = np.zeros([nstep, n])
x_all[0] = x_particle
y_all = np.zeros([nstep, n])
y_all[0] = y_particle

Fx_all = np.zeros([nstep-1,n])
Fx_all[0] = Fx[bin_idx[:,1], bin_idx[:,0]]
Fy_all = np.zeros([nstep-1,n])
Fy_all[0] = Fy[bin_idx[:,1], bin_idx[:,0]]

x_particle += vx*dt
x_all[1] = x_particle
y_particle += vy*dt
y_all[1] = y_particle
vx += Fx[bin_idx[:,1], bin_idx[:,0]]*dt
vy += Fy[bin_idx[:,1], bin_idx[:,0]]*dt


for i in range(nstep-2):
#     x_particle, y_particle, vx, vy = destroy_particles(x_particle, y_particle, vx, vy, xmin, xmax, ymin, ymax) # destroy
    density, xedges, yedges, bin_idx = get_particles_per_bin(x_particle, y_particle, npix, extent)
    x_gridpoints = (xedges[1:]+xedges[:-1])/2
    y_gridpoints = (yedges[1:]+yedges[:-1])/2
    pot = get_potential(density, x_gridpoints, y_gridpoints)
    
    grad_pot = np.gradient(np.abs(pot))
    Fx = grad_pot[1]
    Fy = grad_pot[0]
    
    Fx_all[i+1] = Fx[bin_idx[:,1], bin_idx[:,0]]
    Fy_all[i+1] = Fy[bin_idx[:,1], bin_idx[:,0]]
    
    x_particle += vx*dt
    x_all[i+2] = x_particle
    y_particle += vy*dt
    y_all[i+2] = y_particle
    vx += Fx[bin_idx[:,1], bin_idx[:,0]]*dt
    vy += Fy[bin_idx[:,1], bin_idx[:,0]]*dt
```

```python
c = np.random.rand(n)

fig = plt.figure()
ax = plt.axes(xlim=(xmin,xmax), ylim=(ymin, ymax))
scat = ax.scatter([], [], cmap='rgb')
plt.close()

def init():
    scat.set_offsets([])
    return scat,

def animate(i):
    scat.set_offsets(np.column_stack((x_all[i,:],y_all[i,:])))
    scat.set_array(c)
    return scat,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=nstep, interval=200, blit=True)

rc('animation', html='jshtml')
anim
```
