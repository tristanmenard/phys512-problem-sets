### Problem 1

```python
data = np.loadtxt('WMAP_9yr_combined_power_spectrum.txt')
multipole_index = data[:,0]
power_spectrum = data[:,1]
err = data[:,2]
```

```python
import camb
```

```python
def get_spectrum(pars, lmax=1200):
    H0 = pars[0]
    ombh2 = pars[1]
    omch2 = pars[2]
    tau = pars[3]
    As = pars[4]
    ns = pars[5]
    pars = camb.CAMBparams()
    pars.set_cosmology(H0=H0, ombh2=ombh2, omch2=omch2, mnu=0.06, tau=tau)
    pars.InitPower.set_params(As=As, ns=ns, r=0)
    pars.set_for_lmax(lmax, lens_potential_accuracy=0)
    results = camb.get_results(pars)
    powers = results.get_cmb_power_spectra(pars, CMB_unit='muK')
    cmb = powers['total']
    tt = cmb[:,0]
    return tt
```

```python
pars = [65,0.02,0.1,0.05,2e-9,0.96]
cmb = get_spectrum(pars)
cmb_cut = cmb[2:len(power_spectrum)+2] # keep only the CMB model for the multipole values are in the data (2-1200)
```

```python
plt.figure(figsize=(10,8))
plt.errorbar(multipole_index, power_spectrum, yerr=err, fmt='.', barsabove=True)
plt.plot(multipole_index, cmb_cut)
plt.xlabel(r'Multipole index, $l$')
plt.ylabel('CMB power spectrum')
plt.show()
```

```python
resid = power_spectrum - cmb_cut
chisq = np.sum((resid/err)**2)
print('Chi-squared = '+repr(chisq))
```

```python
def get_spectrum_derivs(pars, delta=1e-8, lmax=1199):
    H0, ombh2, omch2, As, ns = pars
    tau = 0.05
    fun = get_spectrum([H0, ombh2, omch2, tau, As, ns], lmax)
    dH0 = (get_spectrum([H0+delta, ombh2, omch2, tau, As, ns], lmax) - ref)/delta
    dombh2 = (get_spectrum([H0, ombh2+delta, omch2, tau, As, ns], lmax) - ref)/delta
    domch2 = (get_spectrum([H0, ombh2, omch2+delta, tau, As, ns], lmax) - ref)/delta
    dAs = (get_spectrum([H0, ombh2, omch2, tau, As+delta, ns], lmax) - ref)/delta
    dns = (get_spectrum([H0, ombh2, omch2, tau, As, ns+delta], lmax) - ref)/delta
    
    fun = fun[2:]
    dH0 = dH0[2:]
    dombh2 = dombh2[2:]
    domch2 = domch2[2:]
    dAs = dAs[2:]
    dns = dns[2:]
    
    derivs = np.zeros([lmax, len(pars)])
    derivs[:,0] = dH0
    derivs[:,1] = dombh2
    derivs[:,2] = domch2
    derivs[:,3] = dAs
    derivs[:,4] = dns
    return fun, derivs
```

```python
def update_lmbda(lmbda, success):
    if success == False:
        if lmbda == 0:
            lmbda = 1
        else:
            lmbda = 0.5*lmbda
    else:
        if lmbda < 0.2:
            lmbda = 0
        else:
            lmbda = lmbda/2
    return lmbda
```

```python
pars = [65, 0.02, 0.1, 2e-9, 0.96] # H0, ombh2, omch2, As, ns
cmb, derivs = get_spectrum_derivs(pars)

lmbda = 0
chisq_threshold = 0.1
Ninv = np.linalg.inv(np.diag(err))
resid = power_spectrum - cmb
chisq = np.sum((resid/err)**2)
print('Initial guess: chi-squared = '+repr(chisq))

for i in range(30):
    grad = 2*np.dot(derivs.transpose(), np.dot(Ninv, resid))
    curve = 2*np.dot(derivs.transpose(), np.dot(Ninv, derivs))
    curve = curve + lmbda*np.diag(np.diag(curve))
    new_pars = pars + np.dot(np.linalg.inv(curve), grad)
    new_cmb, new_derivs = get_spectrum_derivs(new_pars)
    new_resid = power_spectrum - new_cmb
    new_chisq = np.sum((new_resid/err)**2)
    delta_chisq = new_chisq - chisq
    print('Step '+repr(i+1)+': chi-squared change = '+repr(delta_chisq))
    
    if new_chisq <= chisq:
        resid = new_resid
        chisq = new_chisq
        derivs = new_derivs
        pars = new_pars
        lmbda = update_lmbda(lmbda, success=True)
        print('Updating lambda to '+repr(lmbda)+'.')
        if lmbda == 0 and np.abs(delta_chisq) < chisq_threshold and i > 0:
            print('Stopping after '+repr(i)+' iterations.')
            break
    else:
        lmbda = update_lmbda(lmbda, success=False)
        print('Reject! Updating lambda to '+repr(lmbda)+'.')
```
