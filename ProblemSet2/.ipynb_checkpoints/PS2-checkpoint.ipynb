```python
x = np.linspace(0.5,1,200)
y_true = np.log2(x)
noise = np.random.randn(len(x))
y_data = y_true + 0.1*noise

plt.plot(x, y_data, '.')
plt.plot(x, y_true)
plt.show()
```

```python
order1 = 15
A = np.zeros([len(x), order1+1])
A[:,0] = 1
if order1 > 0:
    A[:,1] = x
    for i in range(2,order1+1):
        A[:,i] = 2*x*A[:,i-1]-A[:,i-2]
```

```python
N = np.eye(len(x))
Ninv = np.linalg.inv(N)
lhs = np.dot(A.transpose(), np.dot(Ninv, A))
rhs = np.dot(A.transpose(), np.dot(Ninv, y_data))
lhs_inv = np.linalg.inv(lhs)
fitp = np.dot(lhs_inv, rhs)
pred = np.dot(A, fitp)
```

```python
plt.plot(x, y_data, '.')
plt.plot(x, y_true)
plt.plot(x, pred)
plt.show()
```
